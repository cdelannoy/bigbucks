```python
# splitting dataset into a training and testing dataset based on date
delta_train = datetime.datetime.strptime('2021-06-19', '%Y-%m-%d') - datetime.datetime.strptime('2021-06-01', '%Y-%m-%d')

train_days = []
for i in range(delta_train.days + 1):
    train_days.append(datetime.datetime.strptime('2021-06-01', '%Y-%m-%d') + datetime.timedelta(days=i))

delta_test = datetime.datetime.strptime('2021-06-30', '%Y-%m-%d') - datetime.datetime.strptime('2021-06-20', '%Y-%m-%d')

test_days = []
for i in range(delta_test.days + 1):
    test_days.append(datetime.datetime.strptime('2021-06-20', '%Y-%m-%d') + datetime.timedelta(days=i))

insiderDat.TradeDate= [datetime.datetime.strptime(d, '%Y-%m-%d %H:%M:%S') for d in insiderDat.TradeDate]

training_data = insiderDat[insiderDat['TradeDate'].isin(train_days)]
testing_data = insiderDat[insiderDat['TradeDate'].isin(test_days)]
```
